```python
y = np.array([35,25,25,15])
mylabels = ['Apple','Banana','cherries','dates']
plt.pie(y, labels = mylabels)
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\ASUS\Downloads\house price prediction.csv")
df['date'] = pd.to_datetime(df['date'])  # Fix dates
df = df.dropna()  # Remove nulls
df_filtered = df[df['price'] > 0]
```

```python
df
```

```python
sns.lineplot(data=df, x='date', y='price')
```

```python
# Scatter plot - axis level function
sns.scatterplot(data = df,x = 'date',y='price')
```

```python
# rel plot - figure level function-square shape
sns.relplot(data=df,x='date',y='price',kind='scatter')
```
