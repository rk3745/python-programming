```python
# (iii) Create a dictionary, do all operations like addition of elements, remove elements, copy items & create nested dictionaries too.
# Step 1: Create a dictionary of students with their marks
students = {"aadesh": 85,"salman": 92,"abhishek": 78,"zohan": 88,"rahul": 90}
print("Original Dictionary:")
print(students)
```

```python
# Step 2: Add new elements
students["amir"] = 95
students.update({"prateek": 80})
print("After Adding Elements:")
print(students)
```

```python
# Step 3: Change (update) an element
students["abhishek"] = 82
print("After Changing Kabir's Marks:")
print(students)
```

```python
# Step 4: Remove elements
students.pop("prateek")
removed_item = students.popitem()
print("After Removing Elements:")
print(students)
print("Removed Item:", removed_item)
```

```python
# Step 5: Copy dictionary
students_copy = students.copy()
print("Copied Dictionary:")
print(students_copy)
```

```python
# Step 6: Create a nested dictionary
classroom = {"ClassA": {"aadesh": 85,"salman": 88},"ClassB": {"abhishek": 95,"zohan": 80}}
print("classroom:",classroom)
```

```python
# Step 7: Access nested dictionary items
print("Accessing Nested Items:")
print("Marks of abhishek in ClassB:", classroom["ClassB"]["abhishek"])
```
