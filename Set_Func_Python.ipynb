***Set = Set items are unordered, unchangeable, and do not allow duplicate values. Sets are used to store multiple items in a single variable. Set items are unchangeable, but you can remove items and add new items.***

***Create Set***

```python
set1 = {1,2,3,4,5,6}
print(set1)
```

***Data Type***

```python
type(set1)
```

***Access Items***

```python
set1 = {1,2,3,4,5,6}
print(4 in set1)
```

```python
set2 = {'a','b','c','d','e'}
print('c' in set2)
```

```python
color = {'Red','Green','Grapes','Oranges'}
print('Pink' in color)
```

```python
color = {'Red','Green','Grapes','Oranges'}
print('Green' in color)
```

***Add Set Item***

```python
set2 = {'a','b','c','d','e'}
set2.add(5)
print(set2)
```

```python
names = {'Alica','Mehul','Chanchal','Alia','Tony'}
names.add(55)
print(names)
```

```python
fruits = {'Apple','Grapes','Orange','Pine apple'}
fruits.add('Guava')
print(fruits)
```

***Update Method***

```python
set1 = {1,2,3,4,5,6}
set2 = {'a','b','c','d','e'}
set1.update(set2)
print(set1)
```

```python
names = {'Alica','Mehul','Chanchal','Alia','Tony'}
fruits = {'Apple','Grapes','Orange','Pine apple'}
names.update(fruits)
print(names)
```

```python
list1 = ['A','B','C']
set1.update(list1)
print(set1)
```

***Remove Method***

```python
set2 = {'a','b','c','d','e'}
set2.remove('a')
print(set2)
```

```python
names = {'Alica','Mehul','Chanchal','Alia','Tony'}
names.remove('Tony')
print(names)
```

```python
num = {45,25,46,78,98,3,4,122,255}
num.remove(25)
print(num)
```

***Discard Method***

***Joining set Union method***

```python
set1 = {1,2,3,4,5,6}
set2 = {'a','b','c','d','e'}
set3 = set1.union(set2)
print(set3)
```

```python
fruits = {'Apple','Grapes','Orange','Pine apple'}
names = {'Alica','Mehul','Chanchal','Alia','Tony'}
n = {45,25,654,6,464,5,4,5,4,24,5,45}
x = fruits.union(names, n)
print(x)
```

```python
set1 = {'A','B','C','D','E'}
set2 = {1,2,3,4,5}
set3 = {'Red','Green','Orange','Pink','Black'}
set4 = {455,789,254,365,456}
set5 = set1.union(set2, set3, set4)
print(set5)
```

***| Operator***

```python
set1 = {1,2,3,4,5,6}
set2 = {'a','b','c','d','e'}
set3 = set1|set2
print(set3)
```

```python
fruits = {'Apple','Grapes','Orange','Pine apple'}
names = {'Alica','Mehul','Chanchal','Alia','Tony'}
n = {45,25,654,6,464,5,4,5,4,24,5,45}
x = fruits|names|n
print(x)
```

```python
set1 = {'A','B','C','D','E'}
set2 = {1,2,3,4,5}
set3 = {'Red','Green','Orange','Pink','Black'}
set4 = {455,789,254,365,456}
x = set1|set2|set3|set4
print(x)
```
