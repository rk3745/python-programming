***Using the Formate Method***

```python
age = 36
txt = ("I am John, and I am {}")
print(txt.format(age))
```

```python
city = 'Delhi'
x = 'I am Alica, and I am from {}'
print(x.format(city))
```

```python
txt = 'I am Vikas, and my Favorite subject {}, and I Living {}'
subject = 'Python'
city = 'Noida'
print(txt.format(subject,city))
```

```python
quantity = 3
itemno = 567
price = 49.95
myorder = "I want {} pieces of item {} for {} dollars."
print(myorder.format(quantity, itemno, price))
```

***Count Method***

```python
txt = 'Welcome to my world, and Welcome to my country'
x = txt.count('o')
print(x)
```

***Join Method***

```python
myTuple = ("John", "Peter", "Vicky")
x = " ".join(myTuple)
print(x)
```
